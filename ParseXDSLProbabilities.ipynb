```python
from bs4 import BeautifulSoup
```

```python

with open("wordle.xdsl", "r") as xdsl:
    dom = BeautifulSoup(xdsl)
```

```python
conditional_probabilities = {
}

for cpt in dom.nodes.findAll("cpt"):
    letters = []
    
    for state in cpt.findAll("state"):
        letters.append(state["id"])
    
    print(cpt["id"], len(letters))
        
    for parent in cpt.findAll("parents"):
        print(parent)
        
    for probabilities in cpt.findAll("probabilities"):
        print(len(probabilities.text.split(" ")))
        
        conditional_probabilities[cpt["id"]] = probabilities.text.split(" ")
        
    print()
```
